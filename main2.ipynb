```python
#Price prediction for master card data
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor, AdaBoostRegressor, BaggingRegressor
from sklearn.linear_model import ElasticNet, Lasso, BayesianRidge, LassoLarsIC, LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load the Mastercard data containing the 'Close' prices
mastercard_data = pd.read_csv('mastercard_close_prices.csv')

# Split the data into features (X) and target variable (y)
X = mastercard_data.drop('Date', axis=1)  # Assuming 'Date' column is not a feature
y = mastercard_data['Mastercard_Close']

# Initialize the models
models = [
    RandomForestRegressor(random_state=42),
    GradientBoostingRegressor(random_state=42),
    ExtraTreesRegressor(random_state=42),
    AdaBoostRegressor(random_state=42),
    BaggingRegressor(random_state=42),
    ElasticNet(),
    Lasso(),
    BayesianRidge(),
    LassoLarsIC(),
    LinearRegression(),
    Ridge(),
    DecisionTreeRegressor(random_state=42),
    KNeighborsRegressor()
]

# Train each model and make predictions
predictions = []

for model in models:
    model.fit(X, y)  # Train on entire dataset
    preds = model.predict(X)
    predictions.append(preds)

# Ensemble by averaging predictions
ensemble_preds = sum(predictions) / len(predictions)

# Calculate the ensemble's mean squared error
ensemble_mse = mean_squared_error(y, ensemble_preds)
print("Ensemble Mean Squared Error:", ensemble_mse)

# Add the ensemble predictions as a new column in the existing Mastercard data
mastercard_data['Predicted_Close'] = ensemble_preds

# Plot the actual and predicted prices
plt.figure(figsize=(10, 6))
plt.plot(mastercard_data['Date'], mastercard_data['Mastercard_Close'], label='Actual Close Price', color='blue')
plt.plot(mastercard_data['Date'], mastercard_data['Predicted_Close'], label='Predicted Close Price', color='green')
plt.title('Actual vs Predicted Close Prices of Mastercard')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from keras.optimizers import Adam
import matplotlib.pyplot as plt

# Load the Visa data containing the 'Close' prices
visa_data = pd.read_csv('visa_close_prices.csv')

# Split the data into features (X) and target variable (y)
X = visa_data.drop('Date', axis=1)  # Assuming 'Date' column is not a feature
y = visa_data['Visa_Close']

# Normalize the data
# Add any necessary feature scaling or normalization here

# Reshape input data for CNN
X_cnn = np.array(X).reshape(X.shape[0], X.shape[1], 1)

# Initialize the CNN model
cnn_model = Sequential()

# Add Convolutional layers
cnn_model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X.shape[1], 1)))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Dropout(0.2))

cnn_model.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Dropout(0.2))

# Flatten the output from convolutional layers
cnn_model.add(Flatten())

# Add Dense layers
cnn_model.add(Dense(units=64, activation='relu'))
cnn_model.add(Dropout(0.2))

cnn_model.add(Dense(units=1))

# Compile the CNN model
cnn_model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Train the CNN model
history = cnn_model.fit(X_cnn, y, epochs=100, batch_size=32, validation_split=0.2)

# Make predictions using the CNN model
cnn_preds = cnn_model.predict(X_cnn)

# Calculate evaluation metrics for the CNN model
mse_cnn = mean_squared_error(y, cnn_preds)
r2_cnn = r2_score(y, cnn_preds)
mae_cnn = mean_absolute_error(y, cnn_preds)

print("CNN Model Evaluation Metrics:")
print("Mean Squared Error:", mse_cnn)
print("R-squared Score:", r2_cnn)
print("Mean Absolute Error:", mae_cnn)

# Plot training and validation loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot actual vs predicted prices for the CNN model
plt.figure(figsize=(10, 6))
plt.plot(visa_data['Date'], y, label='Actual Close Price', color='blue')
plt.plot(visa_data['Date'], cnn_preds, label='Predicted Close Price (CNN)', color='green')
plt.title('Actual vs Predicted Close Prices of Visa (CNN Model)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout, Bidirectional
from keras.optimizers import Adam
import matplotlib.pyplot as plt

# Load the Visa data containing the 'Close' prices
visa_data = pd.read_csv('visa_close_prices.csv')

# Split the data into features (X) and target variable (y)
X = visa_data.drop('Date', axis=1)  # Assuming 'Date' column is not a feature
y = visa_data['Visa_Close']

# Normalize the data
# Add any necessary feature scaling or normalization here

# Reshape input data for LSTM
X_lstm = np.array(X).reshape(X.shape[0], X.shape[1], 1)

# Initialize the LSTM model
lstm_model = Sequential()
lstm_model.add(LSTM(units=500, return_sequences=True, input_shape=(X.shape[1], 1)))
lstm_model.add(Dropout(0.2))
lstm_model.add(LSTM(units=500))
lstm_model.add(Dropout(0.2))
lstm_model.add(Dense(units=1))
lstm_model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Train the LSTM model
lstm_model.fit(X_lstm, y, epochs=100, batch_size=32, validation_split=0.2)

# Make predictions using the LSTM model
lstm_preds = lstm_model.predict(X_lstm)

# Initialize the BiLSTM model
bilstm_model = Sequential()
bilstm_model.add(Bidirectional(LSTM(units=500, return_sequences=True), input_shape=(X.shape[1], 1)))
bilstm_model.add(Dropout(0.2))
bilstm_model.add(Bidirectional(LSTM(units=500)))
bilstm_model.add(Dropout(0.2))
bilstm_model.add(Dense(units=1))
bilstm_model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Train the BiLSTM model
bilstm_model.fit(X_lstm, y, epochs=100, batch_size=32, validation_split=0.2)

# Make predictions using the BiLSTM model
bilstm_preds = bilstm_model.predict(X_lstm)

# Ensemble predictions
ensemble_preds = (lstm_preds.flatten() + bilstm_preds.flatten()) / 2

# Calculate evaluation metrics
rmse = np.sqrt(mean_squared_error(y, ensemble_preds))
mae = mean_absolute_error(y, ensemble_preds)
r2 = r2_score(y, ensemble_preds)

print("Ensemble Model Evaluation Metrics:")
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared Score:", r2)

# Plot actual vs predicted prices
plt.figure(figsize=(10, 6))
plt.plot(visa_data['Date'], y, label='Actual Close Price', color='blue')
plt.plot(visa_data['Date'], ensemble_preds, label='Predicted Close Price (Ensemble)', color='green')
plt.title('Actual vs Predicted Close Prices of Visa (Ensemble Model)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
